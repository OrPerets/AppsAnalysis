```python
import re
import pandas as pd
import numpy as np
import os
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
from sklearn import metrics

print("reading data..")
unsupervised_df = pd.read_csv("learning_data.csv")

unsupervised_df.drop(columns=["Unnamed: 0", "App_Id"],inplace=True)

print("covert to dummies..")
unsupervised_df.drop(unsupervised_df[unsupervised_df['Category'] =='.'].index, inplace=True)
unsupervised_dummy=pd.get_dummies(unsupervised_df)

for col in ["Price","Sentiment","Reviews_Number","Growth (60 days)","Installs","1 Star ratings","5 Star ratings","Thumbs_Up_Count"]:
    unsupervised_dummy[col] = (unsupervised_dummy[col] - unsupervised_dummy[col].mean()) / unsupervised_dummy[col].std()

def get_kmeans_accuracy(data, top_k):
    sum_squared = []
    K = range(2, top_k)
    for i in K:
        print("k=", i, "from", top_k)
        kmeans = KMeans(n_clusters = i, init = 'k-means++')
        kmeans.fit(data)
        sum_squared.append(kmeans.inertia_)         
    return pd.DataFrame({
    "K": K,
    "SSE": sum_squared,
  })

print("Running KMEANS..")
result=get_kmeans_accuracy(unsupervised_dummy,20)  
result.set_index("K", inplace=True)  
result.plot()
```

```python
kmeans = KMeans(n_clusters = 6, init = 'k-means++')
kmeans.fit(unsupervised_dummy)
unsupervised_df["Cluster"] = kmeans.labels_
```
